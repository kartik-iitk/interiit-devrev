```python
from transformers import AutoModelForQuestionAnswering, AutoTokenizer
from optimum.intel.openvino import OVConfig, OVQuantizer

model_id = "deepset/roberta-base-squad2"
model = AutoModelForQuestionAnswering.from_pretrained(model_id)
tokenizer = AutoTokenizer.from_pretrained(model_id)
save_dir = "quantized_roberta"
```

```python
quantization_config = OVConfig()
quantizer = OVQuantizer.from_pretrained(model)
```

```python
def preprocess_function(examples, tokenizer):
  print(examples)
  return tokenizer(examples['question'], examples['context'], return_tensors="pt",padding=True, truncation=True)
```

```python
from functools import partial

calibration_dataset = quantizer.get_calibration_dataset(
    "squad_v2",
    preprocess_function=partial(preprocess_function, tokenizer=tokenizer),
    num_samples=1000,
    dataset_split='train'
)
```

```python
quantizer.quantize(
    quantization_config=quantization_config,
    calibration_dataset=calibration_dataset,
    save_directory=save_dir,
)
```

```python
tokenizer.save_pretrained(save_dir)
```

```python
from functools import partial
from transformers import AutoModelForQuestionAnswering, AutoTokenizer
from neural_compressor.config import PostTrainingQuantConfig
from optimum.intel.neural_compressor import INCQuantizer
from neural_compressor.config import PostTrainingQuantConfig, AccuracyCriterion, TuningCriterion
```

```python
model_id = "deepset/electra-base-squad2"
model = AutoModelForQuestionAnswering.from_pretrained(model_id)
tokenizer = AutoTokenizer.from_pretrained(model_id)
save_dir = "static_electra"
```

```python
!pip install evaluate
```

```python
import evaluate
from datasets import load_dataset
from transformers import pipeline


task_evaluator = evaluate.evaluator("question-answering")
eval_dataset = load_dataset("squad_v2", split="validation").select(range(200))
qa_pipeline = pipeline("question-answering", model=model, tokenizer=tokenizer, handle_impossible_answer=True)
def eval_fn(model):
    qa_pipeline.model = model
    metrics = task_evaluator.compute(model_or_pipeline=qa_pipeline, data=eval_dataset, metric="squad_v2")
    return metrics["f1"]
```

```python
def preprocess_function(examples, tokenizer):
  print(examples)
  return tokenizer(examples['question'], examples['context'], return_tensors="pt",padding=True, truncation=True)
```

```python
accuracy_criterion = AccuracyCriterion(tolerable_loss=0.05)
quantization_config = PostTrainingQuantConfig(approach="static", calibration_sampling_size=[20],
                                        accuracy_criterion=accuracy_criterion)
quantizer = INCQuantizer.from_pretrained(model, eval_fn=eval_fn)
```

```python
calibration_dataset = quantizer.get_calibration_dataset(
    "squad_v2",
    preprocess_function=partial(preprocess_function, tokenizer=tokenizer),
    num_samples=300,
    dataset_split='train'
)
```

```python
quantizer.quantize(
    quantization_config=quantization_config,
    calibration_dataset=calibration_dataset,
    save_directory=save_dir,
)
```
